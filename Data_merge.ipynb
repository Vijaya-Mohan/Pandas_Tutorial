# Task 3: Lookup Operation and Automation (30 points)


Lookup Operation:<br>
<br>
Create a dictionary mapping student_id to exam_score from the scores DataFrame<br>
Use the .map() function to add exam scores to the students DataFrame<br>
Display students with their scores (showing NaN for students without scores)<br>

```python
lookup_dict = df_scores_data.set_index('student_id')['exam_score'].to_dict()
print(lookup_dict)
```

```python
df_students_data_copy = df_students_data.copy()
df_students_data_copy["exam_score"] = df_students_data_copy['student_id'].map(lookup_dict)
print(df_students_data_copy)
```

Performance Comparison:<br>
Implement the same score addition using pd.merge() with a left join <br>

```python
left_join1 = pd.merge(df_students_data, df_scores_data, on = 'student_id', how = 'left')
print(left_join1)
```

Explain why lookup (map) is more efficient than merge for this scenario.<br>
In map() we no need to align indexes or manage duplicate columns and it maps the dictionary directly to the defined Dataframe.<br>
But in merge() we need to build the entire DataFrame which is quite inefficient.

```python
def auto_merge(df1, df2, join_type, key_column):
    if join_type == 'inner':
        merged_df = pd.merge(df1, df2, on=key_column, how='inner')
    elif join_type == 'left':
        merged_df = pd.merge(df1, df2, on=key_column, how='left')
    elif join_type == 'right':
        merged_df = pd.merge(df1, df2, on=key_column, how='right')
    elif join_type == 'outer':
        merged_df = pd.merge(df1, df2, on=key_column, how='outer')
    count = len(merged_df)
    result_dict = {'result_df': merged_df, 'count': count, 'join_type': join_type}
    return result_dict
```

```python
print("=== TASK 3: Lookup and Automation === \n Lookup Operation Result:\n")
input_join_type = input("Enter the type of join (inner, left, right, outer): ")
result = auto_merge(df_students_data, df_enrollments_data, input_join_type, 'student_id')
print(f"The result of the {input_join_type} join is:\n{result}")
```
