```python
data = { 'Name': ['Ali', 'Bobe', 'Charl', 'Davi', 'Ele'],
         'Age': [25, 30, 35, 55, 45],
         'City': ['Chennai', 'Bangalore', 'Hyderabad', 'Delhi', 'Mumbai'] }
pd.DataFrame(data) 
```

```python
raw_data = """Name,Age,City
Ali,25,Chennai
Bobe,30,Bangalore
Charl,35,Hyderabad
Davi,55,Delhi
Ele,45,Mumbai"""
with open("Raw_data.csv", "w")as f:
    f.write(raw_data)

df =pd.read_csv("Raw_data.csv")
df
```
