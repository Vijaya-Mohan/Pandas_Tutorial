```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")
```

# Task 1: Basic Grouping and Single Aggregations

Perform the following grouping operations:<br>
<br>
Calculate total sales amount for each region using groupby() and sum()<br>
Count the number of transactions for each product category using groupby() and count()<br>
Calculate the average sales amount per salesperson using groupby() and mean()<br>
For each result, use reset_index() to convert the result into a clean DataFrame<br>
Sort the regional sales results in descending order using sort_values(ascending=False) to identify the top-performing region<br>

```python
total_sales_amount = df.groupby('region')['sales_amount'].sum()
print('The  total sales amount for each region :\n',total_sales_amount)
number_of_transactions = df.groupby('category')['transaction_id'].count()
print("\nThe Number of transaactions for each product category:\n", number_of_transactions)
average_sales_amount = df.groupby('salesperson')['sales_amount'].mean()
print("\nThe average sales amount per salesperson:", average_sales_amount)
```

```python
total_sales_amount = df.groupby('region')['sales_amount'].sum().reset_index()
print('The  total sales amount for each region :\n',total_sales_amount)
number_of_transactions = df.groupby('category')['transaction_id'].count().reset_index()
print("\nThe Number of transaactions for each product category:\n", number_of_transactions)
average_sales_amount = df.groupby('salesperson')['sales_amount'].mean().reset_index()
print("\nThe average sales amount per salesperson:", average_sales_amount)
```

```python
print("========== Task 1 ==========") 
print("\nBasic Grouping and Single Aggregations\n")
total_sales_amount = df.groupby('region')['sales_amount'].sum().sort_values(ascending=False).reset_index()
print('The  total sales amount for each region :\n',total_sales_amount)
number_of_transactions = df.groupby('category')['transaction_id'].count().sort_values(ascending=False).reset_index()
print("\nThe Number of transaactions for each product category:\n", number_of_transactions)
average_sales_amount = df.groupby('salesperson')['sales_amount'].mean().sort_values(ascending=False).reset_index()
print("\nThe average sales amount per salesperson:", average_sales_amount)
```

# Task 2: Multi-Column Grouping and Multiple Aggregations

```python
total_sales = df.groupby(['region', 'category'])['sales_amount'].sum()
print("Group by both region AND category to calculate total sales for each combination:",total_sales)
```

```python
salesperson_details = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()
print(salesperson_details)
```

```python
total_sales_amount = df.groupby('salesperson')['sales_amount'].sum().sort_values(ascending=False)
print(total_sales_amount)
Top_performer = df.groupby('salesperson')['sales_amount'].sum().idxmax()
print("\nThe Top performer in Sales:",Top_performer)
```

# Custom Aggregation and Complete Sales Report

Define a custom aggregation function that calculates the sales range (max - min) for each group

```python
def sales_range(x):
    return x.max() - x.min()
```

Apply this custom function along with standard aggregations to analyze sales by region:

```python
sales_range_region = df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()
print(sales_range_region)
```

```python
print("========== Final Summary Report ==========\n")
final_summary = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()
print(final_summary)
```

Explain the multi-level column structure that results from the dictionary-based aggregation and how it differs from single aggregations
<br>
applying multiple aggregations (sum, mean) to sales_amount and a single aggregation (count) to customer_id creates a multi-level column structure.<br>
Level 0 (outer): original column names (sales_amount, customer_id)<br>
Level 1 (inner): aggregation functions (sum, mean, count)<br>
This hierarchical format allows multiple results for a single column in one table, making the output more organized and efficient compared to flat, single-aggregation results.
Because of this mix, Pandas organizes the result hierarchically
Multi-level columns improve efficiency by combining multiple results in one structured, computation-friendly format, reducing redundancy and improving clarity for complex analysis.
