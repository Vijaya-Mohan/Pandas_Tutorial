Let's import Pandas library as 'pd'(Though alias naming is optional but it's convinent and great practice! )

```python
import pandas as pd
```

Creating a 2 Dimensional data (Table) using dictionary and saving the data in DataFrame for later use.

```python
students = {"StudentID" : [101,102,103,104,105,106,107,108,109,110],
            "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan", "Fiona", "George", "Hannah", "Ivan", "Julia"],
            "Age": [20,21,19,22,20,21,23,19,22,20],
            "Marks": [88,75,92,65,78,85,70,95,60,80]
            }
df = pd.DataFrame(students)
print(df)
```

Saving the DataFrame to a CSV file named "students.csv" without the index column, here I located the file in the current working directory(You can specify a different path if needed.)

```python
df.to_csv("students.csv", index=False)
```

Let's read the students.csv file and load into a new dataframe.

```python
new_df = pd.read_csv("students.csv")
print(new_df)
```
