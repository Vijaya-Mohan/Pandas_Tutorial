Data Type Conversion

```python
new_df = df.copy()
```

```python

new_df['age'] = pd.to_numeric(new_df['age'], errors = 'coerce')

print(f"Find the data types change in the age column: {new_df.dtypes}")
```

```python
new_df.info()
```

```python
new_df['weight'] = pd.to_numeric(new_df['weight'], errors= 'coerce')
new_df.info()
```

```python
print(f"Number of null values in df:\n{df.isnull().sum()}\n")
print(f"Number of null values in new_df:\n{new_df.isnull().sum()}")
```

```python
Null_count = df['age'].isnull().sum() + df['weight'].isnull().sum()
New_null_count = new_df['age'].isnull().sum() + new_df['weight'].isnull().sum()
print(Null_count)
print(New_null_count)
if Null_count != New_null_count:
    print(f"The number of missing values has changed after converting to numeric :{New_null_count - Null_count}")
```

Tassk 3: Handle Missing Values

```python
new_df['age'] = new_df['age'].fillna(new_df['age'].median())
new_df.head()
```

```python
new_df['weight'] = new_df['weight'].fillna(new_df['weight'].median())
new_df.head()
```

```python
new_df['blood_pressure'] = new_df['blood_pressure'].fillna(new_df['blood_pressure'].median())
new_df.head()
```

```python
new_df['medication'] = new_df['medication'].fillna(new_df['medication'].mode()[0])
new_df.head()
```

```python
new_df['insurance_provider'] = new_df['insurance_provider'].fillna('Unknown')
new_df.head()
```

Task  4: Handle Duplicates

```python
print(new_df[new_df.duplicated(keep = False)])
```

```python
print(new_df.duplicated(subset=['patient_id']))
```

```python
new_df = new_df.drop_duplicates(subset='patient_id', keep='first')
```
